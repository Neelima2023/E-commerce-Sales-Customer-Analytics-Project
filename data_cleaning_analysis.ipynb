```python
import pandas as pd

# Load data
orders = pd.read_csv(r"C:\Users\samba\Documents\Project datasets\olist_orders_dataset.csv")
customers = pd.read_csv(r"C:\Users\samba\Documents\Project datasets\olist_customers_dataset.csv")
payments = pd.read_csv(r"C:\Users\samba\Documents\Project datasets\olist_order_payments_dataset.csv")
items = pd.read_csv(r"C:\Users\samba\Documents\Project datasets\olist_order_items_dataset.csv")

# Merge step by step
df = orders.merge(customers, on="customer_id", how="left")
df = df.merge(payments, on="order_id", how="left")
df = df.merge(items, on="order_id", how="left")

print(df.shape)
df.head()
```

```python
print(df[['order_id', 'customer_id']].head())
print(df.columns)
```

```python
# Total Revenue
total_revenue = df['payment_value'].sum()
print("Total Revenue:", total_revenue)

# Total Orders
total_orders = df['order_id'].nunique()
print("Total Orders:", total_orders)

# Unique Customers
total_customers = df['customer_unique_id'].nunique()
print("Total Customers:", total_customers)
```

```python
# Top States by Revenue
state_sales = df.groupby('customer_state')['payment_value'].sum().sort_values(ascending=False)
print(state_sales.head())

# Monthly Revenue Trend
monthly_sales = df.groupby(['year', 'month'])['payment_value'].sum()
print(monthly_sales.head())

# Top 10 Customers
top_customers = df.groupby('customer_unique_id')['payment_value'].sum().sort_values(ascending=False).head(10)
print(top_customers)
```

```python
df['order_purchase_timestamp'] = pd.to_datetime(df['order_purchase_timestamp'])
```

```python
df['year'] = df['order_purchase_timestamp'].dt.year
df['month'] = df['order_purchase_timestamp'].dt.month
```

```python
monthly_sales = df.groupby(['year', 'month'])['payment_value'].sum()
print(monthly_sales.head())
```

```python
# Top States by Revenue
state_sales = df.groupby('customer_state')['payment_value'].sum().sort_values(ascending=False)
print(state_sales.head())

# Monthly Revenue Trend
monthly_sales = df.groupby(['year', 'month'])['payment_value'].sum()
print(monthly_sales.head())

# Top 10 Customers
top_customers = df.groupby('customer_unique_id')['payment_value'].sum().sort_values(ascending=False).head(10)
print(top_customers)
```

```python
df.to_csv("final_ecommerce_data.csv", index=False)
```
